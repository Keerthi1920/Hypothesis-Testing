**QUESTION 1**

 A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


ANS: Here we have to measure there is any difference in the diameter of the cutlets between two sample units.

This is Two Sample Two Tail test.

Null Hypothesis: Diameter of both the units are same

Alternate Hypothesis: Both units are different from each other


```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv('Cutlets.csv')
df
```

```python
UnitA=df.iloc[:,0]
UnitA
```

```python
UnitB=df.iloc[:,1]
UnitB
```

```python
stats.ttest_ind(UnitA,UnitB)
```

We got the Probability value as 0.47223

Pvalue(0.47223) > level of significance(0.05)

We have to accept the Null Hypothesis

There is no diffference between two units.Both are equal

**QUESTION 2**

A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

Ans: Anova test : Analysis of variance between more than 2 samples.

 Null Hypothesis : All samples TAT population mean are same.
 
 Alternative hypothesis: Atleast one Sample TAT population mean is different.

```python
from google.colab import files
uploaded=files.upload()
```

```python
Lab=pd.read_csv('LabTAT.csv')
Lab.head()
```

```python
from scipy import  stats
stats.f_oneway(Lab.iloc[:,0],Lab.iloc[:,1],Lab.iloc[:,2],Lab.iloc[:,3])
```

Here Probability value  (2..115e-57) < level of significance(0.05).

We are rejecting the Null Hypothesis.

Atleast one sample TAT population mean is different.


**QUESTION 3**

Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.


Ans: Chi-square test

Null Hypothesis: There is no relationship between two categorical variables. The variables are independent.

Alternative Hypothesis: There is a relationship between two categorical variables. They variables are dependent.

```python
from google.colab import files
uploaded=files.upload()
```

```python
BR=pd.read_csv('BuyerRatio.csv')
BR
```

```python
from scipy.stats import chi2_contingency
```

```python
BuyerRatio= [[50,	142,	131,	70],[435,	1523,	1356,	750]]
chi2_contingency(BuyerRatio)
```

Here Probability value(0.66030) > level of significance(0.05)

Accept the Null Hypothesis.

Indpendence of categorical variables. Male-Female buyer ratios are similar across the regions  are not related.
